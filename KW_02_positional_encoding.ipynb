```python
import numpy as np
import matplotlib.pyplot as plt


def get_positional_encoding(N, d_model):
    """
    Generates sinusoidal positional encoding.
    :param N: Number of positions (e.g., maximum sequence length).
    :param d_model: The dimensionality of the model's word embeddings.
    :return: An (N, d_model) numpy array containing the positional encoding for each position.
    """
    position = np.arange(N)[:, np.newaxis]
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))

    positional_encoding = np.zeros((N, d_model))
    positional_encoding[:, 0::2] = np.sin(position * div_term)
    positional_encoding[:, 1::2] = np.cos(position * div_term)

    return positional_encoding


# Generate positional encoding for 80 words, each with dimension of 64
N = 80  # Number of positions
d_model = 64  # Dimensionality of embeddings
positional_encoding = get_positional_encoding(N, d_model)

# Plotting
plt.figure(figsize=(10, 8))
plt.imshow(positional_encoding, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Sinusoidal Positional Encoding")
plt.xlabel("Embedding Dimensions")
plt.ylabel("Word Position")
plt.show()
```
