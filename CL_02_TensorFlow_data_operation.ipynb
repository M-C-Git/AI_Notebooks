```python
import tensorflow as tf
import numpy as np
```

```python
# create a tensor object
t = tf.constant([[1,2,3], [4,5,6]])
print (t)
```

```python
# attributs of tensor object
print(t.shape)
print(t.dtype)
```

```python
# indxing, same as numpy array
print(t[:, 1:])
print(t[...,1,np.newaxis])    # same as t[:,1,np.newaxis]
```

```python
# tensor operation
print(t + 10)
print(tf.square(t))
print(tf.transpose(t))
print(t @ tf.transpose(t))
```

```python
# tensor vs numpy: play well with each other
a = np.array([2,4,5])
tf.constant(a)      # create tensor object from numpy array
t.numpy()           # convert tensor object to numpy array
tf.square(a)        # use tensor funciton to work on numpy array
np.square(t)
```
