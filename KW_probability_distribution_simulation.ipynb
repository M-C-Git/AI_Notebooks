```python
from scipy.stats import gamma, expon, norm, uniform
import numpy as np


np.mean(norm.rvs(loc = 5, scale=2, size=1000)) # simulated expectation of a normal distribution with mean of 5 and std of 2
np.mean(uniform.rvs(loc=5, scale=13-5, size=1000)) #simulated expectation of a continuous distribution distribution with min of 5 and max of 13
```

An exponential distribution has $\lambda =2 $.
Calculate probability of $P(X\le 1)$

```python
print(expon.cdf(1, scale = 1/2)) # Analytical
n=1000000
m = expon.rvs(scale=1/2, size=n)
np.mean(m)
print(np.sum(m<1)/n) #simulation based
```

```python
# Generating 100 sets of random numbers based on exponential distribution using scipy

import matplotlib.pyplot as plt

m = expon.rvs(scale=1/2, size=1000000)

plt.hist(m, bins=30)
plt.show()
```
